```python
# import torch
import sys, os
sys.path.append('../')
import matplotlib.pyplot as plt
import seaborn as sns

import json
import pandas as pd
import numpy as np
import glob
```

```python
save_path = '/gpfs/data/geraslab/Nan/saves/selfsup/swav_breast/data_20220111_full'

model_path = 'swav_resnet18_avgpool_coslr-100e_largebatch_skynet-gpu32'
file = '20220330_115900.log.json'
f = open(os.path.join(save_path, model_path, file))
lines = f.readlines()[1:]
logs_batch_vanilla = pd.DataFrame([json.loads(line) for line in lines])

model_path = 'swav_resnet18_milatten1_batch128'
file = "20220427_060949.log.json"
f = open(os.path.join(save_path, model_path, file))
lines = f.readlines()[1:]
logs_batch_milatt = pd.DataFrame([json.loads(line) for line in lines])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(logs_batch_vanilla.index*128, logs_batch_vanilla.loss, label='vanilla ', alpha=0.6)
plt.plot(logs_batch_milatt.index*128, logs_batch_milatt.loss, label='attention', alpha=0.6)
plt.xlabel('#mini-batches x batchsize')
plt.ylabel('SwAV loss')
plt.legend()
```
